```python
from typing import TypedDict, List
from langgraph.graph import StateGraph, START, END
```

```python
class AgentState(TypedDict):
    number1: int
    number2: int
    operation1: str
    number3: int
    number4: int
    operation2: str
    answer1: int
    answer2: int

def adder1(state: AgentState) -> AgentState:
    """string for ai"""

    state["answer1"] = state["number1"] + state["number2"]
    return state

def subtractor1(state: AgentState) -> AgentState:
    """string for ai"""

    state["answer1"] = state["number1"] - state["number2"]
    return state

def adder2(state: AgentState) -> AgentState:
    """string for ai"""

    state["answer2"] = state["number3"] + state["number4"]
    return state

def subtractor2(state: AgentState) -> AgentState:
    """string for ai"""

    state["answer2"] = state["number3"] - state["number4"]
    return state

def decide_next_node1(state: AgentState) -> AgentState:
    """string for ai"""

    if state["operation1"] == "+":
        return "addition_operation"
    
    if state["operation1"] == "-":
        return "subtraction_operation"
    
def decide_next_node2(state: AgentState) -> AgentState:
    """string for ai"""

    if state["operation2"] == "+":
        return "addition_operation"
    
    if state["operation2"] == "-":
        return "subtraction_operation"

```

```python
graph = StateGraph(AgentState)

graph.add_node("add_node1", adder1)
graph.add_node("subtract_node1", subtractor1)
graph.add_node("conditional_edge1", lambda state: state)
graph.add_node("add_node2", adder2)
graph.add_node("subtract_node2", subtractor2)
graph.add_node("conditional_edge2", lambda state: state)


graph.add_edge(START, "conditional_edge1")

graph.add_conditional_edges(
    "conditional_edge1", #source node
    decide_next_node1, #conditional function

    {
        #edge: node
        "addition_operation": "add_node1",
        "subtraction_operation": "subtract_node1"
    }
)

graph.add_edge("add_node1", "conditional_edge2")
graph.add_edge("subtract_node1", "conditional_edge2")

graph.add_conditional_edges(
    "conditional_edge2",
    decide_next_node2,

    {
        "addition_operation": "add_node2",
        "subtraction_operation": "subtract_node2"
    }
)

graph.add_edge("add_node2", END)
graph.add_edge("subtract_node2", END)

app = graph.compile()
```

```python
result = app.invoke({"number1": 1, "number2": 2, "operation1": "-", "number3": 5, "number4": 7, "operation2": "+"})

result["answer1"]
```
