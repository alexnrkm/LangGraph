```python
from langgraph.graph import StateGraph, START, END
from typing import TypedDict, Dict, List
import random
```

```python
class AgentState(TypedDict):
    name: str
    numbers: List[int]
    counter: int
```

```python
def greeting_node(state: AgentState) -> AgentState:
    state["name"] = f"Hey {state['name']}"
    state["counter"] = 0

    return state

def random_node(state: AgentState) -> AgentState:
    state["numbers"].append(random.randint(0, 10))
    state["counter"] += 1

    return state

def should_continue(state:AgentState) -> AgentState:
    if state["counter"] < 5:
        return "continue"
    else:
        return "exit"
```

```python
graph = StateGraph(AgentState)

graph.add_node("greeting_node", greeting_node)
graph.add_node("random_node", random_node)

graph.add_edge(START, "greeting_node")
graph.add_edge("greeting_node", "random_node")

graph.add_conditional_edges(
    "random_node", #source node
    should_continue, #condition function

    {
        "continue": "random_node", #if continue edge to the random_node
        "exit": END #if exit edge to the end
    }

)
```

```python
app = graph.compile()

app.invoke({"name": "Alex", "numbers": [], "counter": 0})
```
