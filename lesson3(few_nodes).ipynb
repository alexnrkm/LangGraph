```python
from typing import TypedDict, List
from langgraph.graph import StateGraph
```

```python
class AgentState(TypedDict):
    name: str
    age: int
    skills: List[str]
    message: str

def greeting_node(state: AgentState) -> AgentState:
    state["message"] = f"Hello {state['name']}! "

    return state

def age_node(state:AgentState) -> AgentState:
    state["message"] = state["message"] + f"Your age is {state['age']}, "

    return state

def skills_node(state:AgentState) -> AgentState:
    state["message"] = state["message"] + f"Your skills are {', '.join(state['skills'])}"

    return state
```

```python
graph = StateGraph(AgentState)

graph.add_node("first_node", greeting_node)
graph.add_node("second_node", age_node)
graph.add_node("third_node", skills_node)

graph.set_entry_point("first_node")
graph.add_edge("first_node", "second_node")
graph.add_edge("second_node", "third_node")
graph.set_finish_point("third_node")

app = graph.compile()
```

```python
result = app.invoke({"name": "Alex", "age": 20, "skills": ["SQL", "Python", "LangGraph"]})

result["message"]
```
